```python
from machinevisiontoolbox import mkcube
import matplotlib.pyplot as plt
X,Y,Z = mkcube(edge=True)  # cube of side length 1
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_wireframe(X,Y,Z)
```

```python
from machinevisiontoolbox import CentralCamera
cam = CentralCamera(f=0.015, rho=10e-6, imagesize=[1280, 1024], pp=[640, 512], name="mycamera")
print(cam)
```

```python
from spatialmath import SE3
X1, Y1, Z1 = mkcube(1, pose=SE3(0.1, 0.2, 3), edge=True)


cam.plot_wireframe(X1,Y1,Z1)
```
